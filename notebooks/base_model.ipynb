```python
import sys
import os
sys.path.append(os.path.abspath(".."))
import torch
from torch.utils.data import Dataset, DataLoader, SubsetRandomSampler,random_split, Subset 
import torch.nn as nn
import wfdb
import numpy as np
from os import listdir
from os.path import exists
import matplotlib.pyplot as plt
import torchviz
import torch.nn.functional as F
from tqdm import tqdm
from torch.optim.lr_scheduler import StepLR
import time
import torch.optim as optim
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from torch.optim import Adam
from src.data.loading import loading_and_filtering_
from src.data.preprocessing import Preprocessor
from src.models.wnet1d import WNet1D
from src.models.loss_func import CombinedSignalLoss
from src.training.trainer import Trainer
```

```python
#loading data and filtering valid records only
database_path = "../bidmc"
records_file_path = f"{database_path}/RECORDS"
required_signal = ["PLETH,", "II,"]
sampling_freq = 125
min_duration = 8  # in minutes

 #Step 1: Load valid record paths--
valid_records = loading_and_filtering_(database_path=database_path,
    file_path=records_file_path,
    min_duration=min_duration,
    required_signal=required_signal,
    sampling_freq=sampling_freq
)

# Hyperparameters
batch_size = 128
learning_rate = 1e-3
num_epochs = 500
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)

# Initialize your model (make sure to instantiate it)
loss_fn = CombinedSignalLoss()
rmse_list = []
pearson_list = []
results_dir = "../results/base_model"
os.makedirs(results_dir, exist_ok=True)
for record in valid_records:
    print(f"Record {record}")
    dataset = Preprocessor(record)
    # Step 1: Split into train and validation
    train_size = int(0.8 * len(dataset))
    indices = list(range(len(dataset)))
    val_size = len(dataset) - train_size
    #train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
    train_indices = indices[:train_size]#using the last 20% as validation to avoid information leakage from segmentasion
    val_indices   = indices[train_size:]
    train_dataset = Subset(dataset, train_indices)
    val_dataset   = Subset(dataset, val_indices)

    # Step 2: Create DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    model = WNet1D().to(device)
    trainer = Trainer(model=model, train_loader=train_loader, val_loader=val_loader, loss_fn=loss_fn, lr=1e-3,
                 optimizer_class=Adam, scheduler_class=StepLR, scheduler_kwargs={"step_size": 800, "gamma": 0.1},
                 device=device)
    # Step 4: Train
    history=trainer.train(num_epochs=num_epochs )
    # Step 5: Evaluate with metrics
    metrics = trainer.evaluate()  
    print(f"Record {record} --> RMSE: {metrics['rmse']:.4f}, Pearson r: {metrics['pearson_r']:.4f}") 
    rmse_list.append(metrics["rmse"])            # <<< Added
    pearson_list.append(metrics["pearson_r"]) 
    # Step 6: Save model
    model_name = f"base_model_{record.replace('/', '_')}.pt"
    model_path = os.path.join(results_dir, model_name)
    trainer.save_model(model_path)
    print(f"Saved model to {model_path}")

    # Step 7: Plot and save loss curves 
    plt.figure(figsize=(10, 5))
    plt.plot(history["train_loss"], label="Train Loss")
    plt.plot(history["val_loss"], label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Training Curve for Record {record}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    figure_name = f"base_model_{record.replace('/', '_')}.png"
    figure_path = os.path.join(results_dir, figure_name)
    plt.savefig(figure_path)
    print(f"Saved training plot to {figure_path}")
    plt.show()
```

```python
# Calculate means
mean_rmse = np.mean(rmse_list)
mean_pearson = np.mean(pearson_list)

np.save("rmse_list_base_model_Alon.npy", rmse_list)
np.save("pearson_list_base_model_Alon.npy", pearson_list)

# Print means
print(f"Mean RMSE: {mean_rmse:.4f}")
print(f"Mean Pearson r: {mean_pearson:.4f}")

# Plot RMSE histogram
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(rmse_list, bins=15, color='skyblue', edgecolor='black')
plt.axvline(mean_rmse, color='red', linestyle='--', label=f'Mean = {mean_rmse:.4f}')
plt.title("RMSE Histogram")
plt.xlabel("RMSE")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)

# Plot Pearson r histogram
plt.subplot(1, 2, 2)
plt.hist(pearson_list, bins=15, color='lightgreen', edgecolor='black')
plt.axvline(mean_pearson, color='red', linestyle='--', label=f'Mean = {mean_pearson:.4f}')
plt.title("Pearson Correlation Histogram")
plt.xlabel("Pearson r")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
save_dir = os.path.join("..", "results","base_model")
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "RMSE_and_r_base_model_Alon.png")
plt.savefig(save_path, dpi=300)
plt.show()
```
